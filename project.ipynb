### code

```python
import pandas as pd
```

```python
eduDf = pd.read_csv("API_SE.XPD.TOTL.GD.ZS_DS2_en_csv_v2_11528.csv")
lifeDf = pd.read_csv("API_SP.DYN.LE00.IN_DS2_en_csv_v2_11567.csv")
healthDf = pd.read_csv("public-health-expenditure-share-gdp.csv")
```

```python
eduDf.head()
```

```python
lifeDf.head()
```

```python
healthDf.head()
```

```python
eduDf_2019 = eduDf[["Country Name", "2019"]].copy()
eduDf_2019 = eduDf_2019.rename(columns={
    "Country Name": "Country",
    "2019": "EducationSpending"
})
eduDf_2019.head()
```

```python
lifeDf_2019 = lifeDf[["Country Name", "2019"]].copy()
lifeDf_2019 = lifeDf_2019.rename(columns={
    "Country Name": "Country",
    "2019": "LifeExpectancy"
})
lifeDf_2019.head()
```

```python
healthDf_2019 = healthDf[healthDf["Year"] == 2019][
    ["Entity", "Public health expenditure as a share of GDP"]
].copy()

healthDf_2019 = healthDf_2019.rename(columns={
    "Entity": "Country",
    "Public health expenditure as a share of GDP": "HealthSpending"
})
healthDf_2019.head()
```

```python
df = eduDf_2019.merge(lifeDf_2019, on="Country") \
               .merge(healthDf_2019, on="Country")
df = df.dropna()
```

```python
df.head()
```

#### linear regression

```python
X = df[["EducationSpending", "HealthSpending"]]
y = df["LifeExpectancy"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
coef_df = pd.DataFrame({
    "Variable": X.columns,
    "Coefficient": model.coef_
})

intercept = model.intercept_

coef_df, intercept
```

LifeExpectancyi​ = β0​ + β1​ х EducationSpendingi ​+ β2​ х HealthSpendingi ​+ εi

LifeExpectancy ​= 72.8 ​+ 0.14 х EducationSpending + 1.006 х HealthSpending

intercept represents the theoretical life expectancy when all explanatory variables are equal to zero. in this situation it is equal to 72-73 years.

```python
r2 = model.score(X, y)
r2
```

The coefficient of determination R² equals 0.38, which means that approximately 38% of the variation in life expectancy across countries is explained by education and healthcare expenditures included in the model.

Multicollinearity refers to a situation where independent variables in a regression model are highly correlated with each other, which can make coefficient estimates unstable and difficult to interpret.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df[["EducationSpending", "HealthSpending"]].dropna()

vif = pd.DataFrame()
vif["Variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif
```

```python
X = df[["EducationSpending", "HealthSpending"]]

X.corr()
```

```python
import matplotlib.pyplot as plt

# fitted values (predictions)
fitted = model.predict(X)

# residuals
residuals = y - fitted

plt.scatter(fitted, residuals)
plt.axhline(0, linestyle="--")
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")
plt.show()
```

The residuals versus fitted values plot does not show a clear funnel-shaped pattern, suggesting no strong visual evidence of heteroscedasticity. However, several outliers are present, which may affect inference.

```python
import statsmodels.api as sm

X_sm = sm.add_constant(X)
ols_model = sm.OLS(y, X_sm).fit()
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan
het_breuschpagan(ols_model.resid, ols_model.model.exog)
```

The presence of heteroscedasticity was tested using the Breusch–Pagan test.
The test yielded a p-value of 0.996, indicating that the null hypothesis of homoscedasticity cannot be rejected at the 5% significance level.

#### log-linear / log-log regression

LifeExpectancy = β0 ​+ β1​ln(EducationSpending) + β2​ln(HealthSpending) + ε

```python
(df[["EducationSpending", "HealthSpending"]] <= 0).sum()
```

```python
import numpy as np

df["log_EducationSpending"] = np.log(df["EducationSpending"])
df["log_HealthSpending"] = np.log(df["HealthSpending"])
```

```python
X_log = df[["log_EducationSpending", "log_HealthSpending"]]
X_log = sm.add_constant(X_log)

y = df["LifeExpectancy"]

log_model = sm.OLS(y, X_log).fit()
print(log_model.summary())
```

The coefficient on education spending is negative but statistically insignificant, suggesting no clear evidence of an association with life expectancy in the cross-sectional sample for 2019.

encreasing health spending by 1% encreases life expectancy by 0.0628 years (≈23 days). formula:
Δy≈β×0.01

```python
bp_test = het_breuschpagan(log_model.resid, log_model.model.exog)
bp_test
```

#### robust regression

LifeExpectancy = β0 ​+ β1​ln(EducationSpending) + β2​ln(HealthSpending) + ε

we are using the same as previous one to ensure a correct comparison

```python
from sklearn.linear_model import HuberRegressor

X_robust = df[["log_EducationSpending", "log_HealthSpending"]]
y = df["LifeExpectancy"]

huber = HuberRegressor()
huber.fit(X_robust, y)

huber.coef_, huber.intercept_
```

```python
robust_results = pd.DataFrame({
    "Coefficient": ["Intercept", "log_EducationSpending", "log_HealthSpending"],
    "Estimate": np.concatenate(([huber.intercept_], huber.coef_))
})

robust_results
```

The results confirm a strong positive association between health expenditure and life expectancy, while education spending remains statistically insignificant.
